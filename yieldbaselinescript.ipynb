```python
import pandas as pd

# Load the CSV file
file_path = '/content/ICRISAT-District Level Data.csv'
data = pd.read_csv(file_path)

# Print the column names
print("Column names in the CSV file:")
print(data.columns.tolist())

# Define the yield columns
yield_columns = [
    'RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)', 'KHARIF SORGHUM YIELD (Kg per ha)',
    'RABI SORGHUM YIELD (Kg per ha)', 'SORGHUM YIELD (Kg per ha)', 'PEARL MILLET YIELD (Kg per ha)',
    'MAIZE YIELD (Kg per ha)', 'FINGER MILLET YIELD (Kg per ha)', 'BARLEY YIELD (Kg per ha)',
    'CHICKPEA YIELD (Kg per ha)', 'PIGEONPEA YIELD (Kg per ha)', 'MINOR PULSES YIELD (Kg per ha)',
    'GROUNDNUT YIELD (Kg per ha)', 'SESAMUM YIELD (Kg per ha)', 'RAPESEED AND MUSTARD YIELD (Kg per ha)',
    'SAFFLOWER YIELD (Kg per ha)', 'CASTOR YIELD (Kg per ha)', 'LINSEED YIELD (Kg per ha)',
    'SUNFLOWER YIELD (Kg per ha)', 'SOYABEAN YIELD (Kg per ha)', 'OILSEEDS YIELD (Kg per ha)',
    'SUGARCANE YIELD (Kg per ha)', 'COTTON YIELD (Kg per ha)'
]

# Filter the data for years between 1966 and 2017 and exclude 0 and -1 yield values
filtered_data = data[(data['Year'] >= 1966) & (data['Year'] <= 2017)]
filtered_data = filtered_data.replace({0: pd.NA, -1: pd.NA})

# Calculate the mean yield for each district and crop
def calculate_mean_yield(df, yield_columns):
    df_grouped = df.groupby(['State Name', 'Dist Name'])
    yield_sum = df_grouped[yield_columns].sum(min_count=1)  # Use min_count to keep NaNs
    valid_years_count = df_grouped[yield_columns].count()
    mean_yields = yield_sum.div(valid_years_count.where(valid_years_count != 0, pd.NA))
    return mean_yields

average_yields = calculate_mean_yield(filtered_data, yield_columns)

# Save the result to a CSV file
output_file_path = '/content/average_yields.csv'
average_yields.to_csv(output_file_path)

print("Average yields for each crop for each district from 1966 to 2017:")
print(average_yields)

print(f"\nThe result has been saved to {output_file_path}")
```
